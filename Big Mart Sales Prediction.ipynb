Handline missing values

1.   Mean -> Avg Value
2.   Mode -> Most Repeated Value



```python
big_mart_data['Item_Weight'].mean()
```

Filling the missing values with mean for the Item_Weight

```python
big_mart_data['Item_Weight'].fillna(big_mart_data['Item_Weight'].mean(), inplace=True)
```

```python
big_mart_data.isnull().sum()
```

Filling the missing values with mean for the Outlet_Size

```python
mode_of_outlet_size = big_mart_data.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
mode_of_outlet_size
```

```python
missing_values = big_mart_data['Outlet_Size'].isnull()
missing_values
```

```python
big_mart_data.loc[missing_values, 'Outlet_Size'] = big_mart_data.loc[missing_values, 'Outlet_Type'].apply(lambda x: mode_of_outlet_size.loc['Outlet_Size', x])
```

Data Preprocessing

```python
big_mart_data.head()
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

```python
big_mart_data.replace({
    'Item_Fat_Content': {'low fat':'Low Fat', 'LF':'Low Fat', 'reg':'Regular'}
}, inplace=True)
```

```python
big_mart_data['Item_Fat_Content'].value_counts()
```

Label Encoding

```python
encoder = LabelEncoder()
```

```python
 #Item_Identifier
 #Item_Fat_Content
 #Item_Type
 #Outlet_Identifier
 #Outlet_Size
 #Outlet_Location_Type
 #Outlet_Type
big_mart_data['Item_Identifier'] = encoder.fit_transform(big_mart_data['Item_Identifier'])
big_mart_data['Item_Fat_Content'] = encoder.fit_transform(big_mart_data['Item_Fat_Content'])
big_mart_data['Item_Type'] = encoder.fit_transform(big_mart_data['Item_Type'])
big_mart_data['Outlet_Identifier'] = encoder.fit_transform(big_mart_data['Outlet_Identifier'])
big_mart_data['Outlet_Size'] = encoder.fit_transform(big_mart_data['Outlet_Size'])
big_mart_data['Outlet_Location_Type'] = encoder.fit_transform(big_mart_data['Outlet_Location_Type'])
big_mart_data['Outlet_Type'] = encoder.fit_transform(big_mart_data['Outlet_Type'])
```

```python
big_mart_data.head()
```

Splitting the features and target

```python
x = big_mart_data.drop(['Item_Outlet_Sales'], axis=1)
y = big_mart_data['Item_Outlet_Sales']
```

Splitting the data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
x.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

machine learning model training, XGBoost Regressor

```python
regressor = XGBRegressor()
regressor.fit(x_train, y_train)
```

Evaluation

```python
#prediction on training data
training_data_prediction = regressor.predict(x_train)

#R squared value
r2_train = metrics.r2_score(y_train, training_data_prediction)
r2_train # if it is close to 1 it is good
```

```python
#prediction for testing data
test_data_prediction = regressor.predict(x_test)

#R squared value
r2_test = metrics.r2_score(y_test, test_data_prediction)
r2_test
```

```python
input_data = x_test.iloc[1]

predictions = regressor.predict([input_data])
predictions
```
